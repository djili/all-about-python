```python
import matplotlib.pyplot as plt
import numpy as np 
```

## Tracer un diagrame a point 

```python
# creation des vecteurs de donnees x (valeurs 30 reguliere entre 0 et 50)
x = np.linspace(0, 50, 30)
# creation des vecteurs de donnees x (valeurs 30 reguliere entre 0 et 50)
y = np.random.uniform(0, 30, 30 )
```

```python
print(x)
```

```python
print(y)
```

```python
plt.figure()
plt.scatter(x, y)
plt.show()
```

## Diagramme a barre 

```python
x = np.arange(10) 
y = np.random.uniform(1, 10, 10)
```

```python
print(x)
```

```python
print(y)
```

```python

plt.figure()
plt.bar(x,y)
plt.show()
```

## Diagramme pixel

```python
z = np.random.uniform(0.0,10, (32,32))
```

```python
print(z)
```

```python
plt.figure()
plt.imshow(z)
plt.show()
```

## Diagramme heatmap

```python
plt.figure()
plt.contourf(z)
plt.show()
```

## HeatDiagramme heatmap v2

```python
plt.figure()
plt.contour(z)
plt.show()
```
