```python
# Conversion des colonnes en float
dataframe['Source Value'] = dataframe['Source Value'].str.replace(',', '.').astype(float)
dataframe['Reading'] = dataframe['Reading'].str.replace(',', '.').astype(float)
dataframe.info()
```

```python
import matplotlib.pyplot as plt

# Create figure and axis with specific size
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(dataframe['Source Value'], dataframe['Reading'], 'r-', label='Reading', alpha=0.7)

# Customize the plot
ax.set_title('Evolution of Reading over Source Value')
ax.set_xlabel('Source Value')
ax.set_ylabel('Reading')
ax.grid(True, linestyle='--', alpha=0.6)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend
ax.legend()

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()

plt.savefig("../output")
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
model  = LinearRegression()

model.fit(dataframe['Source Value'].values.reshape(-1, 1), dataframe['Reading'])
```

```python
x_line = np.linspace(-0.7, 0.7, 100)
y_line = model.predict(x_line.reshape(-1, 1))
```

```python

# Create figure and axis with specific size
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(dataframe['Source Value'], dataframe['Reading'], 'r-', label='Reading', alpha=0.7)
ax.plot(x_line, y_line, '-g', alpha=0.7, label='Linear curve')

# Customize the plot
ax.set_title('Evolution of Reading over Source Value')
ax.set_xlabel('Source Value')
ax.set_ylabel('Reading')
ax.grid(True, linestyle='--', alpha=0.6)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend
ax.legend()

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```
